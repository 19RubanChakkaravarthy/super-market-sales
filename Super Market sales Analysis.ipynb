### HANDLING OUTLIERS

```python
def replace_outliers_grouped(df, group_col, num_col):
    g = df.groupby(group_col)[num_col]
    Q1 = g.transform(lambda s: s.quantile(0.25))
    Q3 = g.transform(lambda s: s.quantile(0.75))
    IQR = Q3 - Q1
    upper = Q3 + 1.5 * IQR
    lower = Q1 - 1.5 * IQR

    mask_high = df[num_col] > upper
    mask_low = df[num_col] < lower
    df.loc[mask_high, num_col] = upper[mask_high]
    df.loc[mask_low, num_col] = lower[mask_low]
    return df
```

```python
df = replace_outliers_grouped(df, group_col="Branch", num_col="Total")
```

```python
sns.boxplot(
    x="Branch",
    y="Total",
    hue="Branch",
    data=df,
    palette="Set3",
    notch=True,
    linewidth=2,
    dodge=False,           # overlap boxes so legend entries aren't duplicated
    legend=False
)
```

```python
df = replace_outliers_grouped(df, group_col="Branch", num_col="Profit")
```
